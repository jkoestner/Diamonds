## James Allen Diamond Parser

This code scrapes the James Allen website for available diamonds and ouputs them to a database as well as an excel file.

*This website is changed frequently so the beautiful soup code may not parse the code correctly if changed. It is advised to modify code to handle new source code if needed.*


```python
#importing the necessary libraries

import pandas as pd
import urllib3
import certifi
import os
import time
from bs4 import BeautifulSoup
from collections import namedtuple
from json import loads
import requests
import json
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)
```

```python
#defining the http parameter

http = urllib3.PoolManager(
    cert_reqs='CERT_REQUIRED', # Force certificate check.
    ca_certs=certifi.where(),  # Path to the Certifi bundle.
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/51.0.2704.84 Safari/537.36'}
)   
```

```python
#going through the pages to output the results of diamonds

http = urllib3.PoolManager()
df = pd.DataFrame()   
df.empty
colors = ['F', 'G', 'H', 'I']
for co in colors:
    x=0
    err=0
    while err <1:
        x+=1
        url = (f"https://www.jamesallen.com/loose-diamonds/emerald-cut/page-{x}/?Color={co}&Clarity=VS2,VS1,VVS2,VVS1&CaratFrom=1&CaratTo=1.9&resultsView=List")
        print(url)
        response = http.request('GET', url)
        soup = BeautifulSoup(response.data)
        data_page_container = soup.find(id="end-of-page") #no longer needed but still could be useful
        data_page_container_string = str(soup)
        a = data_page_container_string.find('[{"DiamondID')
        if a == -1:
            err = 1
        else:
            b = data_page_container_string.find('}]},')
            output = data_page_container_string[a:b+2]
            parsed_json = json.loads(output)
            diamond=pd.DataFrame(parsed_json)
            df = df.append(diamond)
df.shape
```

```python
#outputting the dataframe to csv file

df.to_csv('out.csv')
```
